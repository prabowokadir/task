2. Dari folder `data_tugas1` baca data `avocado.json` di dalam pandas dan simpan dataframe kedalam `dataframe_2` lalu tampilkan tabelnya. Kemudian save kedalam `avocado.csv`.

3. Dari folder `data_tugas1` baca data `chinook.db` dengan `sqlite3`. Di dalam pandas ambil tabel kolom `Employeeid`,`LastName`,`Title` dari tabel `employees` dan tampilkan tabelnya. Simpan tabel kedalam `dataframe_3`, kemudian save kedalam `employees.json` dengan orientasi `columns`. 

```python
import sqlite3
```

```python
### Tulis kode anda disini
con = sqlite3.connect('data_tugas1/chinook.db')
dataframe_3 = pd.read_sql('select Employeeid, LastName, Title from employees', con)
dataframe_3
```

```python
dataframe_3.to_json('employees.json', orient='columns')
```

4. Bacalah file `avocado.json` kedalam variabel `avocado`. Lalu urutkan data avocado terhadap tanggal, pada method sort tersebut tambahkan argument `ignore_index=True`. Simpan hasil akhir dalam variabel `dataframe_4`

```python
avocado = pd.read_json('data_tugas1/avocado.json', orient='split') # Baca file avocado.json
avocado = avocado.drop('Unnamed: 0',axis=1) # Jangan ubah ini
dataframe_4 = avocado.sort_values(by='Date', ignore_index=True)# Sort value berdasarkan tanggal, tambahkan arguments ignore_index=True
dataframe_4
```

6. Di dalam dataframe `avocado`, hitunglah nilai *nonzero* pada kolom `XLarge Bags`

7. Pada dataframe `avocado`, hitunglah jumlah data yang memiliki tipe `organic`
