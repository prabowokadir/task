1. Dari folder `data_tugas2`, bacalah file `hr_data.csv` dan tampilkan 5 data teratas dari dataframe tersebut. 

```python
import numpy as np
import pandas as pd

dataframe1 = pd.read_csv('data_tugas2/hr_data.csv', sep=';')
dataframe1.head()
```

2. Dari `dataframe1`, terdapat kolom bernama `sales`, gantilah nama kolom tersebut menjadi `department`. Simpan `DataFrame` yang diedit dalam variabel `dataframe2`, dan tampilkan 5 data teratas. [hint](https://note.nkmk.me/en/python-pandas-dataframe-rename/#:~:text=You%20can%20use%20the%20rename,any%20row%20%2F%20column%20name%20individually.&text=Specify%20the%20original%20name%20and,is%20for%20the%20columns%20name.)

```python
dataframe2 = dataframe1.rename(columns={'sales':'department'})
dataframe2.head()
```

3. Dari `dataframe2`, tampilkanlah kolom `satisfaction_level` dan `department` untuk index ke 20 sampai 35

```python
dataframe2.iloc[20:36, [0, 8]]
```

4. Dengan menggunakan `dataframe2`, ambillah data yang memiliki nilai 0 saja pada kolom `left`, lalu _assign_ kedalam variabel `dataframe3`. Lalu cek info dari dataframe tersebut

```python
dataframe3 = dataframe2[(dataframe2['left']==0)]
dataframe3.info()
```

5. Dengan menggunakan `dataframe2`, ambillah data yang memiliki nilai 1 saja pada kolom `left`, lalu _assign_ kedalam variabel `dataframe4`. Lalu cek info dari dataframe tersebut.

```python
dataframe4 = dataframe2[(dataframe2['left']==1)]
dataframe4.info()
```

6. Dari `dataframe2`, cek jumlah duplikat yang ada pada dalam dataframe. Jika ada duplikat, buang duplikat tersebut dan assign kedalam `dataframe_non_duplicated`.

```python
print('Jumlah data terduplikasi =', dataframe2.duplicated().value_counts()[1])
dataframe_non_duplicated = dataframe2.drop_duplicates()
print('Shape data sebelum hapus duplikat :', dataframe2.shape)
print('Shape data setelah hapus duplikat :', dataframe_non_duplicated.shape)
```
