1. Buatlah kode dengan pandas untuk concatenate df1 dan df2, lalu merge lah hasil concatenation tersebut dengan df3 dengan kolom id sebagai kuncinya. Lalu tampilkan dataframe akhirnya.

```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

df2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})

df3 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
        'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]})

concatenated = pd.concat([df1, df2])
merged_data = pd.merge(concatenated, df3, on='student_id')
merged_data
```

2. Buatlah program pandas dengan left join untuk menggabungkan 2 dataset berikut dengan menggunakan `key1` dan `key2` dari `data1` saja. Lalu tampilkan hasilnya.

```python
data1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})
merged_data = pd.merge(data1, data2, how='left', on=['key1', 'key2'])
merged_data
```

5. Buatlah kode untuk mengelompokkan `student_data` menjadi group berdasarkan kode sekolah.

```python
student_data = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_Of_Birth ': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'age': [12, 12, 13, 13, 14, 12],
    'height': [173, 192, 186, 167, 151, 159],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']},
    index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6'])

result = student_data.groupby(by='school_code')

for name,group in result:
    print("\nGroup:")
    print(name)
    print(group)
```

7. Dari data `SaleData.xlsx`, buatlah perintah untuk pivot table dan carilah total PENJUALAN untuk tiap region dan tiap manager.

```python
sale = pd.read_excel('data_tugas/SaleData.xlsx')
sale.head()
```

```python
sale_amount = sale.pivot_table(index=['Region'], columns=['Manager'], values=['Sale_amt'], aggfunc='sum').stack()
sale_amount
```
