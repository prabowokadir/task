### Tugas
___Load_ `tugas.csv` pada folder `data`__. Jadikan kolom `datetime` sebagai _index_

```python
### jawaban anda disini
import pandas as pd
import numpy as np

data = pd.read_csv('data/tugas.csv', index_col='datetime', parse_dates=True)
data.head()
```

1. Dengan `map`, ubah nilai pada kolom berikut: 
 * `season`: 1 menjadi "_summer_", 2 menjadi "_autumn_", 3 menjadi "_winter_", 4 menjadi "_rainy_"
 * `holiday`: 0 untuk "_non-holiday_", 1 untuk "_holiday_"
 * `workingday`: 0 untuk "_breakday_", 1 untuk "_workingday_"
 * `weather`: 1 menjadi "_clear_", 2 menjadi "_cloudy_", 3 menjadi "_rainy_", 4 menjadi "_stormy_"

```python
### jawaban anda disini
data['season'] = data['season'].map({1:'summer', 2:'autumn', 3:'winter', 4:'rainy'})
data['holiday'] = data['holiday'].map({0:'non-holiday', 1:'holiday'})
data['workingday'] = data['workingday'].map({0:'breakday', 1:'workingday'})
data['weather'] = data['weather'].map({1:'clear', 2:'cloudy', 3:'rainy', 4:'stormy'})
data.head()
```

2. Dengan `applymap` dibantu dengan _slicing_ kolom, ubah seluruh nilai numerik dengan nilai kuadratnya menggunakan lambda

```python
### jawaban anda disini
data.loc[:, 'temp':] = data.loc[:,'temp':].applymap(lambda x: x**2)
data.head()
```

3. Gunakan fungsi _squareroot_ dari Numpy `np.sqrt` untuk mengubah semua data numerik dengan `apply`

```python
### jawaban anda disini
data.loc[:, 'temp':] = data.loc[:,'temp':].apply(np.sqrt)
data.head()
```

4. _Aggregate_ `data` berdasarkan `season`, `holiday`, dan `weather` untuk mendapatkan nilai maksimum dan minimum setiap kolom

```python
### jawaban anda disini
data.groupby(['season', 'holiday', 'weather']).agg([np.max, np.min])
```

5. _Aggregate_ `data` berdasarkan `season`, dan `weather` dengan fungsi berikut, dan ubah nama kolomnya:
 * _Mean_ untuk `temp` menjadi `mean_temperature`
 * _Min_ dan _max_ untuk `count` menjadi `minimal` dan `maximal`
 * Total kuadrat dari `humidity` dengan fungsi sendiri atau lambda, menjadi `humidity peak`

```python
# Function Total Square
def total_square(series):
    total = 0
    for i in series:
        total+=i**2
    return total
```

```python
### jawaban anda disini
new_data = data.groupby(['season', 'weather']).agg({'temp':np.mean, 'count':[np.min, np.max], 'humidity':total_square})
new_data.columns = ['mean_temperature', 'minimal', 'maximal', 'humidity_peak']
new_data
```

6. Kelompokkan data berdasarkan semua kolom kategorik. Kemudian, untuk tiap kolom numerik, transformasikan semua data sehingga nilainya berubah menjadi rasio antara mean dan median. (Boleh membuat _function_ atau menggunakan _lambda_)

```python
### jawaban anda disini
data.groupby(['season', 'holiday', 'workingday', 'weather']).transform(lambda x: x.mean()/x.median())
```

7. **Load `sp500.csv`** pada folder data, dan set kolom `date` sebagai Index. Lalu tampilkan 10 data pertama

```python
### jawaban anda disini
sp500 = pd.read_csv('data/sp500.csv', index_col='date', parse_dates=True)
sp500.head(10)
```

8. Dari data sp500 tadi, lakukanlah _downsampling_ untuk data per minggu dengan fungsi agregasi mean.

```python
import matplotlib.pyplot as plt

### jawaban anda disini
sp500_weekly = sp500.resample('W').mean()

# Plot daily and weekly resampled time series together
start, end = '2013-01', '2013-07'
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(sp500.loc[start:end],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(sp500_weekly.loc[start:end],
marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.legend();
```

9.Lakukanlah _centered rolling window mean_ untuk 7 dan 30 hari pada data `sp500`.

```python
### jawaban anda disini
sp500_7d = sp500.rolling(7, center=True).mean()
sp500_30d = sp500.rolling(30, center=True).mean()

# Start and end of the date range to extract
start, end = '2016-01', '2016-07'
# Plot daily, weekly resampled, and 7-day rolling mean time series together
fig, ax = plt.subplots(figsize=(15,4))
ax.plot(sp500.loc[start:end],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(sp500_7d.loc[start:end],
marker='.', linestyle='-', label='7-d Rolling Mean')
ax.plot(sp500_30d.loc[start:end],
marker='.', linestyle='-', label='30-d Rolling Mean')
ax.legend();
```
